```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
from producecfg import *
from libSOGA import *

from time import time
```

```python
def bubblesort(dim=3):
    vec = norm.rvs(loc=0,scale=1,size=dim)
    #print('Initial array:', vec)
    swap = 0
    for _ in range(len(vec)):
        for i in range(len(vec)-1):
            if vec[i] > vec[i+1]:
                t = vec[i+1]
                vec[i+1] = vec[i]
                vec[i] = t
                swap += 1
    return vec, swap
            
```

# BubbleSort4

```python
samples4 = []

for _ in range(int(1e4)):
    _, swap = bubblesort(dim=4)
    samples4.append(swap)
    
print(np.mean(samples4), np.std(samples4))
```

```python
start = time()
cfg = produce_cfg('Performance/Bubble4')
print(cfg.ID_list)
output_dist = start_SOGA(cfg)
end = time()

print('Computed in:', end-start)
```

```python
output_dist.gm.mean()[4], np.sqrt(output_dist.gm.cov()[4,4])
```
